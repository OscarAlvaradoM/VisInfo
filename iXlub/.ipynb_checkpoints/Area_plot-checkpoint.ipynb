```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Objeto 'Figura'

```python
class Figura:
    """Clase que crea la figura principal, el lienzo en el que se dibujará
    Atributos: 
        ax (Ejes o arreglo de ejes): Objeto o arreglo de objetos tipo 'axis' en donde se puede graficar
        con diferentes tipos de gráficos.
        
        matriz_obj (matriz de objetos tipo plot): Matriz que se está creando para almacenar las gráficas que se requieran empaquetados como 
        objetos
        """
    def __init__(self, n_x_m, alto_x_ancho = [3,3], titulo = 'Título principal'):
        """ Se crea la figura deseada con las dimensiones que se indiquen.
        Parámetros:
            n_x_m (tupla de enteros): Reenglones x columnas de la matriz de ejes (ax) que se está creando
        
            alto_x_ancho (tupla de enteros): Define la separación que habrá entre las figuras plasmadas
            en el lienzo. La separación se mantiene para todas las figuras.
            
            titulo (str): Titulo principal, se mostrará por encima de todas las subplots que se hagan.
            """
        # Separamos la tupla de dimensiones en dos entes por separado.
        self.n = n_x_m[0]
        self.m = n_x_m[1]
        # Separamos la tupla de cuánto se alejan cada subplot del otro respecto a una altura y anchura
        h_pad = alto_x_ancho[0]
        w_pad = alto_x_ancho[1]

        # Inicializamos la figura desde matplotlib con un tamaño especial para usar en Jupyter. Se coloca el
        # título establecido desde un principio y a una altura considerable para no intervenir con lso gráficos.
        fig = plt.figure(figsize = (10,6))
        fig.suptitle(titulo, y = 1.05)
        
        # Se crean los ejes deseados, al igual que se impone la distancia entre subplots.
        self.ax = fig.subplots(self.n, self.m)
        fig.tight_layout(h_pad = h_pad, w_pad = w_pad)
        
        # Creamos el objeto que nos permitirá mannipular a los gráficos de manera cómoda.
        if self.n == 1 and self.m == 1:
            self.matriz_obj = 0
        else:
            self.matriz_obj = np.zeros((self.n,self.m), dtype = "object")
            
            
    def plot_area(self, coord, obj_graf, x, y, **kwargs):
        """Se plotea la gráfica deseada introduciendo el tipo de gráfico que se quiere en forma de objeto. 
        Parámetros:
            coord (tupla de enteros): Indica en qué lugar del arreglo de ejes se quiere graficar.
            
            obj_graf (Objeto): Se invoca el objeto que contiene la información de la gráfica deseada.
            
            x (arreglo 1D): Contiene los puntos que se graficarán, como variable independiente.
            
            y (arreglo ND): Contiene  una matriz con los puntos que se graficarán. Tiene dominio x.
        
            """
        # Se arregla el problema que surge al intentar invocar una figura con un solo subplot o de 2x1 o 1x2.
        self.coord = coord
        if self.n == 1 and self.m == 1:
            self.matriz_obj = obj_graf(self.ax, x, y)
        elif self.n == 1 or self.m == 1:
            self.ax.shape = (self.n, self.m)
            self.matriz_obj[self.coord] = obj_graf(self.ax[self.coord], x, y)
        else:
            self.matriz_obj[self.coord] = obj_graf(self.ax[self.coord], x, y)
        # Se regresa el objeto Figura como tal, ya que hasta este punto no se grafica aún. 
        return self
    
    
    def show(self):
        """Hace un recorrido sobre todos los ejes creados para graficar en dado caso de que se haya dado la 
        orden
            """
        # Ejecuta un for sobre nuestro arreglo de objetos.
        for _,matriz in np.ndenumerate(self.matriz_obj):
            # No hace nada si no se ha puesto nada en determinado sector de la figura
            if matriz == 0:
                pass
            else:
            # Invoca el método dibujar() que está en la clase de la gráfica que se quiera dibujar.
                matriz.dibujar()
                
    def set_axes(self,coord, **kwargs):
        """Sirve para editar los atributos de cada eje (ax) de la Figura. Se utiliza un diccionario como entrada
        Parámetros:
            kwargs (diccionario): La notación de ** obliga al usuario a usar un diccionario e introducirlo con los
            dos asteriscos para poder desempaquedar y usar. Debe contener al menos las llaves 'xlabel', 'ylabel',
            'xlim', 'ylim', 'grif', 'ticks' (diccionario de diccionarios) y 'spines'(otro diccionario de 
            diccionarios).
            """
        # Aquí se toma el ente kwargs y se asigna a una variables del objeto en cuestión
        self.diccionario = kwargs
        # Se usan las características introducidas en el diccionario.
        self.ax[coord].set_xlabel(self.diccionario['xlabel'])
        self.ax[coord].set_ylabel(self.diccionario['ylabel'])
        self.ax[coord].set_xlim(self.diccionario['xlim'])
        self.ax[coord].set_ylim(self.diccionario['ylim'])
        self.ax[coord].grid('grid')
        self.ax[coord].set_xticks(self.diccionario['ticks']['xtick_major'])
        self.ax[coord].set_yticks(self.diccionario['ticks']['ytick_major'])
        self.ax[coord].set_xticks(self.diccionario['ticks']['xtick_minor'], minor = True)
        self.ax[coord].set_yticks(self.diccionario['ticks']['ytick_minor'], minor = True)

        self.ax[coord].spines['right'].set_visible(self.diccionario['spines']['right'])
        self.ax[coord].spines['left'].set_visible(self.diccionario['spines']['left'])
        self.ax[coord].spines['top'].set_visible(self.diccionario['spines']['top'])
        self.ax[coord].spines['bottom'].set_visible(self.diccionario['spines']['bottom'])
        
        #ax[i].set_xticklabels(diccionario[i]['xticks_major_l'])
        #ax[i].set_yticklabels(diccionario[i]['yticks_major_l'])
        #ax[i].set_xticklabels(diccionario[i]['xticks_minor_l'], minor = True)
        #ax[i].set_yticklabels(diccionario[i]['yticks_minor_l'], minor = True)        
```

## Objeto 'Area_plot'

```python
class Area_plot:
    """Clase que crea un objeto tipo figura y que su finalidad es dibujar una gráfica de área.
        """
    def __init__(self, axes, x, y):
        """Se toma en consideración el eje sobre el que se dibujará el gráfico y lo que se graficará
        Parámetros:
            ax (Eje): Un eje (Axis) desde matplotlib para poder graficar sobre este.
            
            x (Arreglo 1D): Puntos para graficar sobre el eje x
            
            y (Arrelgo ND): Puntos para graficar sobre el eje y, necesita tener la misma dimensión 'horizontal'
            para poder graficar contra x
            """
        # Simplemente se toman los parámetros de la clase y los volvemos Atributos
        self.ax = axes
        self.x = x
        self.y = y
              
        
    def dibujar(self):
        """Método para dibujar la gráfica de área o stack area. El segundo es lo mismo que el primero pero 
        generalizado a más de una gráfica."""
        self.ax.stackplot(self.x, self.y)
        
```

### Ejemplo de funcionamiento 1

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

y = np.vstack([y1, y2, y3])

a = Figura([1, 1], titulo = 'Título')
a.plot_area((0,0), Area_plot, x, y)
a.show()
```

### Ejemplo de funcionamiento 2

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x
y = np.vstack([y1, y2, y3])

a = Figura([2, 1], titulo = 'Título')
a.plot_area((0,0), Area_plot, x, np.abs(np.sin(x)))
a.plot_area((1,0), Area_plot, x, np.tan(y))

a.show()
```

### Ejemplo de funcionamiento 3

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x
y = np.vstack([y1, y2, y3])

a = Figura([2, 2], titulo = 'Título')
a.plot_area((0,0), Area_plot, x, np.abs(np.sin(x)))
a.plot_area((1,1), Area_plot, x, np.sin(y))
a.plot_area((0,1), Area_plot, x, np.cos(y))
a.plot_area((1,0), Area_plot, x, np.tan(y))

ticks1 = {'xtick_major': np.linspace(0, 2*np.pi, 10), 'xtick_minor': np.linspace(0.5, 2.5, 0.5), 
          'ytick_major': np.linspace(-0.5, 1, 10), 'ytick_minor': np.linspace(0.5, 2.5, 0.5)}
spines1 = {'right': True, 'left': True, 'top': True, 'bottom': True}

car = {'xlabel': 'x', 'ylabel': 'y', 'xlim': (0, 2*np.pi), 'ylim': (-0.5, 1), 
       'grid': True, 'ticks': ticks1, 'spines': spines1} 

a.set_axes((0,0), **car)

a.show()
```
